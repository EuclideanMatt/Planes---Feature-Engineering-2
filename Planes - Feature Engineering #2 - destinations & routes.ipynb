```python
### Deconstructing the 'Route' feature ###
```

```python
dfc['route'].unique()
```

```python
# let's remove the arrow first

dfc['route'] = dfc['route'].str.replace(r'\s*→\s*', ' ', regex=True)
dfc
```

```python
## splitting all locations by the space and creating new features

dfc['start'] = dfc['route'].str.split(' ').str[0]
dfc['stop_1'] = dfc['route'].str.split(' ').str[1]
dfc['stop_2'] = dfc['route'].str.split(' ').str[2]
dfc['stop_3'] = dfc['route'].str.split(' ').str[3]
dfc['stop_4'] = dfc['route'].str.split(' ').str[4]
dfc['stop_5'] = dfc['route'].str.split(' ').str[5]
```

```python
dfc['start'] = dfc['start'].astype(str)
sorted(dfc['start'].unique())
```

```python
## With the goal to numerate all the categorical features, I want to have the factors throughout the newly created variables be unique
## Example: del will be key 3 through all six new features
```

```python
## This can be accomplished by utilizing dictionaries
```

```python
# creating the key values for the first location

unique_sorted_keys = sorted(dfc['start'].unique())
numerated_dict = {value: idx for idx, value in enumerate(unique_sorted_keys)}

print(numerated_dict)
```

```python
# sorting the rest of the locations

unique_sorted_values_stop_1 = sorted(dfc['stop_1'].unique())
unique_sorted_values_stop_2 = sorted(dfc['stop_2'].unique())
unique_sorted_values_stop_3 = sorted(dfc['stop_3'].unique())
unique_sorted_values_stop_4 = sorted(dfc['stop_4'].unique())
unique_sorted_values_stop_5 = sorted(dfc['stop_5'].unique())
```

```python
# converting the data value to string
dfc['stop_1'] = dfc['stop_1'].astype(str)
dfc['stop_2'] = dfc['stop_2'].astype(str)
dfc['stop_3'] = dfc['stop_3'].astype(str)
dfc['stop_4'] = dfc['stop_4'].astype(str)
dfc['stop_5'] = dfc['stop_5'].astype(str)
```

```python
unique_sorted_values_stop_1 = sorted(dfc['stop_1'].unique())
unique_sorted_values_stop_2 = sorted(dfc['stop_2'].unique())
unique_sorted_values_stop_3 = sorted(dfc['stop_3'].unique())
unique_sorted_values_stop_4 = sorted(dfc['stop_4'].unique())
unique_sorted_values_stop_5 = sorted(dfc['stop_5'].unique())
```

```python
# combining all the key values

combined_unique_values = sorted(set(unique_sorted_keys + unique_sorted_values_stop_1 + unique_sorted_values_stop_2 + unique_sorted_values_stop_3 + unique_sorted_values_stop_4  + unique_sorted_values_stop_5))
```

```python
# adding uniquness between the locations

initial_dict = numerated_dict.copy()
updated_dict = initial_dict.copy()

next_index = max(initial_dict.values()) + 1
for value in combined_unique_values:
    if value not in updated_dict:
        updated_dict[value] = next_index
        next_index += 1

print("Initial dictionary:", [initial_dict])
print("Updated dictionary:", [updated_dict])
```

```python
locations = updated_dict
locations_as_lists = {key: [value] for key, value in locations.items()}
locations_as_lists
```

```python
# encoding our route features 

encoding_dict = updated_dict

dfc['start'] = dfc['start'].map(encoding_dict)
dfc['stop_1'] = dfc['stop_1'].map(encoding_dict)
dfc['stop_2'] = dfc['stop_2'].map(encoding_dict)
dfc['stop_3'] = dfc['stop_3'].map(encoding_dict)
dfc['stop_4'] = dfc['stop_4'].map(encoding_dict)
dfc['stop_5'] = dfc['stop_5'].map(encoding_dict)

dfc
```

```python
dfc.drop('route', axis=1, inplace=True)
dfc.info()
```

```python
### Deconstructing the 'source' & 'destination' features ###
# We will want these on the same scale as our routes
# 1.) abbvreviate 
# 2.) enumerate with updated_dict from above.
```

```python
dfc['source'].unique()
```

```python
source_dict = {
    'banglore': 'blr',
    'kolkata': 'ccu', 
    'delhi': 'del' ,
    'chennai': 'maa',
    'mumbai': 'bom'
}
```

```python
dfc['destination'].unique()
```

```python
 destination_dict = {
    'new delhi': 'del',
    'bangalore': 'blr',
    'cochin': 'cok',
    'kolkata': 'ccu',
    'delhi': 'del',
    'hyderabad': 'hyd'
}
```

```python
# Function to update features using the dictionaries and handle NaN values
def update_features(dfc, source_dict, destination_dict, updated_dict):
    # Update source
    dfc['source'] = dfc['source'].map(source_dict).map(updated_dict)
    # Update destination
    dfc['destination'] = dfc['destination'].map(destination_dict).map(updated_dict)
    
    # Replace NaN values with 0 using assignment to avoid the warning
    dfc['source'] = dfc['source'].fillna(0)
    dfc['destination'] = dfc['destination'].fillna(0)
    
    return dfc

# Assuming df is your dataframe containing 'source' and 'destination' columns
dfc = update_features(dfc, source_dict, destination_dict, updated_dict)

# Display the first few rows to verify the changes
dfc.head()
```
